# Obtenção dos enunciados

```python
import pandas as pd

df_doc_hits = pd.read_csv('../../data/log_juris_tcu/doc-hits.csv')[:50][['ID']]
df_doc_hits
```

```python
df_jurisprudencia_selecionada = pd.read_csv('../../data/juris_tcu/doc.csv')[['ID', 'TEXT']]
df_jurisprudencia_selecionada
```

```python
df = df_doc_hits.merge(df_jurisprudencia_selecionada, on='ID')
df = df.set_index('ID')
df
```

```python
enunciados = df.to_dict()['TEXT']
enunciados
```

# Execução

```python
responses = {}
print('Generating queries:')
for i, id in enumerate(enunciados.keys()):
    print('\tStatement', str(i) + '...')
    texto_enunciado = enunciados[id]
    prompt = f'Elabore até 5 perguntas curtas e diretas que possam ser respondidas a partir do enunciado a seguir: \n\n{texto_enunciado}'
    conversation = ConversationChain(llm=chat, verbose=False)
    response = conversation.predict(input=prompt)
    #print(response)
    responses[id] = response.split('\n')
print('Done')
```

```python
with open('../../data/log_juris_tcu/llm_query.txt', 'w') as f_out:
    for i, id in enumerate(responses.keys()):
        f_out.write('#' + str(id) + '\n')
        f_out.write(enunciados[id] + '\n')
        for query in responses[id]:
            f_out.write(query + '\n')
        f_out.write('---------------------\n')
    
```
