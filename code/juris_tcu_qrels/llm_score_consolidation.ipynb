```python
import pandas as pd
```

# Carga dos scorings

```python
df_most_relevants = pd.read_csv('../../data/llm_juris_tcu/eval_most_relevants.csv')
df_most_relevants
```

```python
print('Média de avaliações por query:', sum(df_most_relevants['LLM_SCORE'].value_counts() / 150), '\nPor Score:')
df_most_relevants['LLM_SCORE'].value_counts() / 150
```

```python
df_least_relevants = pd.read_csv('../../data/llm_juris_tcu/eval_least_relevants.csv')
df_least_relevants
```

# Consolidação

```python
df_run = pd.concat([df_most_relevants, df_least_relevants], ignore_index=True)
df_run = df_run[['QUERY_ID', 'QUERY_TEXT', 'DOC_ID', 'DOC_TEXT', 'LLM_SCORE', 'LLM_REASONING', 'RANK', 'ENGINE']]
df_run = df_run.sort_values(by=['QUERY_ID', 'LLM_SCORE', 'RANK', 'DOC_ID'], ascending=[True, False, True, True])
df_run['ENGINE'] = df_run['ENGINE'] + '+LLM'
df_run = df_run.rename(columns={'LLM_SCORE': 'SCORE'})
df_run['SCORE'] = df_run['SCORE'].astype('int32')
df_run = df_run.reset_index(drop=True)
df_run
```

```python
df_run.query('QUERY_ID < 51')[['QUERY_ID', 'DOC_ID', 'SCORE', 'ENGINE', 'RANK']].to_csv('../../data/juris_tcu/qrel1.csv', index=False)
df_run.query('QUERY_ID > 50 and QUERY_ID < 101')[['QUERY_ID', 'DOC_ID', 'SCORE', 'ENGINE', 'RANK']].to_csv('../../data/juris_tcu/qrel2.csv', index=False)
df_run.query('QUERY_ID > 100')[['QUERY_ID', 'DOC_ID', 'SCORE', 'ENGINE', 'RANK']].to_csv('../../data/juris_tcu/qrel3.csv', index=False)
```

```python
with open('../../data/llm_juris_tcu/evaluations.txt', 'w', encoding='Utf-8') as f_out:
    for idx, row in df_run.iterrows():
        if idx % 15 == 0:
            f_out.write(f"Query {row['QUERY_ID']} - {row['QUERY_TEXT']}\n")
        f_out.write(f"\tEnunciado {row['DOC_ID']} - {row['DOC_TEXT']}\n\t\t\t(rank {row['RANK']} do pipeline {row['ENGINE']})\n\t\tScore {row['SCORE']} - Razão: {row['LLM_REASONING']}\n")       
```
