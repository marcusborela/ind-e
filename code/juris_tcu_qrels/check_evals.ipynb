```python
import pandas as pd
```

```python
df_most_relevants = pd.read_csv('../../data/llm_juris_tcu/eval_most_relevants.csv')
df_most_relevants
```

```python
df_stats = df_most_relevants.groupby(['QUERY_ID', 'LLM_SCORE']).count()[['DOC_ID']]
df_stats.columns = ['COUNT']
df_stats = df_stats.reset_index()
df_stats = df_stats.set_index('QUERY_ID')
df_stats = df_stats.astype({'LLM_SCORE': 'int32'})
df_stats
```

```python
df_stats['SCORE_3'] = df_stats.query('LLM_SCORE == 3')['COUNT']
df_stats['SCORE_2'] = df_stats.query('LLM_SCORE == 2')['COUNT']
df_stats['SCORE_1'] = df_stats.query('LLM_SCORE == 1')['COUNT']
df_stats['SCORE_0'] = df_stats.query('LLM_SCORE == 0')['COUNT']
df_stats = df_stats.fillna('0')
del df_stats['LLM_SCORE']
del df_stats['COUNT']
df_stats = df_stats.astype({'SCORE_3': 'int32', 'SCORE_2': 'int32', 'SCORE_1': 'int32', 'SCORE_0': 'int32'})
df_stats = df_stats.drop_duplicates()
df_stats['TOTAL'] = df_stats['SCORE_3'] + df_stats['SCORE_2'] + df_stats['SCORE_1'] + df_stats['SCORE_0'] 
df_stats
```

```python
df_stats.to_csv('../../data/llm_juris_tcu/eval_statistics.csv')
```
